BMI = weight (kg) ÷ height (m)²

Underweight: < 18.5

Healthy weight: 18.5 – 24.9

Overweight: 25.0 – 29.9

Obesity (Class 1): 30.0 – 34.9

Obesity (Class 2): 35.0 – 39.9

Obesity (Class 3 / Severe Obesity): ≥ 40.0

```python
height = float(input("Enter your height in metres : "))
weight = float(input("Enter your weight in kilograms : "))

bmi = weight / (height ** 2)

if bmi < 18.5:
  print(f"Your BMI is {bmi} which means you are underweight")
elif bmi < 25:
  print(f"Your BMI is {bmi} which means you are healthy weight")
elif bmi < 30:
  print(f"Your BMI is {bmi} which means you are overweight")
elif bmi < 35:
  print(f"Your BMI is {bmi} which means you are obese (Class 1)")
elif bmi < 40:
  print(f"Your BMI is {bmi} which means you are obese (Class 2)")
else:
  print(f"Your BMI is {bmi} which means you are severely obese")
```

```python
import random

def roll_dice(dice):
  return random.choice(dice)

def ladders(player_position):
  ladders_dict = {
      8: 29,
      10: 32,
      20: 39,
      27: 85,
      51: 67,
      72: 91,
      79: 100
  }

  return ladders_dict.get(player_position)

def snakes(player_position):
  snakes_dict = {
      17: 7,
      53: 50,
      56: 24,
      62: 19,
      64: 60,
      93: 73,
      94: 75,
      98: 65
  }

  return snakes_dict.get(player_position)

dice = [1, 2, 3, 4, 5, 6]
snakes_list = [17, 53, 56, 62, 64, 93, 94, 98]
ladders_list = [8, 10, 20, 27, 51, 72, 79]
player_position = 1


while True:
  player_input = input("Press Enter to Play or Exit to Quit : ")
  if player_input.lower() == "exit":
    break
  if player_input == "":
    landing_face = roll_dice(dice)
    # landing_face = int(input("Enter the dice face : "))
    player_position += landing_face
    if player_position in snakes_list:
      player_position = snakes(player_position)
    elif player_position in ladders_list:
      player_position = ladders(player_position)
    if player_position >= 100:
      print("=================")
      print("Hooray! You Won!")
      print("=================")
      break
    else:
      print(f"You landed on {landing_face}, your new position is {player_position}")
```
